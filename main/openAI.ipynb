```python
import requests
import json

url = "https://api.openai-hk.com/v1/chat/completions"

headers = {
    "Content-Type": "application/json",
    "Authorization": "hk-k8uxev10000468101dda6e2b13df9fb77af5c78f05867bdf"
}

response = requests.post(url, headers=headers, data=json.dumps(data).encode('utf-8') )
result = response.content.decode("utf-8")

# 定义生成剧情的函数
def generate_dialog_sequence(word_list):
    # 初始角色和场景的设置
    scenes = []
    last_npc = "智慧老人"  # 初始场景的NPC角色
    
    # 循环生成多个场景
    for i, word in enumerate(word_list):
        # 设定生成对话的提示语
        data = {
    "max_tokens": 1200,
    "model": "gpt-3.5-turbo",
    "temperature": 0.8,
    "top_p": 1,
    "presence_penalty": 1,
    "messages": [
        {
            "role": "system",
            "content": "You are ChatGPT, a large language model trained by OpenAI."
        },
        {
            "role": "user",
            "content": f"""
        在一个神秘的世界里，主角与不同的角色展开对话。每段对话剧情会根据前一个角色的内容进行延续。
        生成第{len(scenes) + 1}个场景的对话，其中包含单词 "{word}"。
        角色: {last_npc}
        对话要根据角色的背景和主角与角色之间的互动进行构建。

        角色将会根据上下文展开对话，使用适当的背景和NPC的个性，确保剧情连贯。

        请返回包含以下信息的对话:
        1. 角色名字
        2. 角色对话内容
        3. 角色头像
        4. 背景视频
        5. 相关的词汇和选择（该词汇将在对话中使用）
        """
        }
    ]
}

        response = requests.post(url, headers=headers, data=json.dumps(data).encode('utf-8') )
        result = response.content.decode("utf-8")
        dialog = json.loads(result)
        
        
        # 解析生成的对话文本
        dialog = response.choices[0].text.strip()

        # 分析对话文本并构造JSON数据结构
        npc_name = f"NPC{len(scenes) + 1}"
        npc_dialog = dialog.strip()
        npc_avatar = f"/avatars/{npc_name}.png"
        background_video = f"/videos/{npc_name.lower()}.mp4"
        vocabulary = {
            "word": word,
            "options": [
                { "text": "正确选项1", "isCorrect": True, "nextScene": len(scenes) + 1 },
                { "text": "错误选项1", "isCorrect": False, "nextScene": 0 },
                { "text": "错误选项2", "isCorrect": False, "nextScene": 0 }
            ]
        }

        # 添加当前场景到列表
        scenes.append({
            "npcName": npc_name,
            "npcDialog": npc_dialog,
            "npcAvatar": npc_avatar,
            "backgroundVideo": background_video,
            "vocabulary": vocabulary
        })

        # 更新下一个NPC的名称
        last_npc = f"NPC{len(scenes) + 1}"

    return json.dumps(scenes, ensure_ascii=False, indent=4)

# 调用函数生成包含大量场景的对话
word_list = ["mysterious", "ambitious", "increasingly", "adventurous", "resilient", "elusive"] * 100  # 根据需要生成的场景数调整
dialog_json = generate_dialog_sequence(word_list)
print(dialog_json)
```
