```python
import requests
import socket

def check_proxy_connection(host='127.0.0.1', ports=[7890, 1080, 8080, 10809]):
    for port in ports:
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        result = sock.connect_ex((host, port))
        if result == 0:
            print(f"端口 {port} 开放")
        else:
            print(f"端口 {port} 关闭")
        sock.close()

# 测试本地代理端口
print("正在测试代理端口...")
check_proxy_connection()

# 测试网络连接
def test_connection():
    test_urls = [
        ('http://www.google.com', '谷歌'),
        ('https://api.openai.com', 'OpenAI API')
    ]
    
    for url, name in test_urls:
        try:
            response = requests.get(url, timeout=5)
            print(f"{name} 连接成功: {response.status_code}")
        except Exception as e:
            print(f"{name} 连接失败: {str(e)}")

print("\n正在测试网络连接...")
test_connection()
```
