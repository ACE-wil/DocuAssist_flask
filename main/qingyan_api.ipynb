````python
import time


word = ['revolution', 'trial', 'amazing', 'fantastic',]

for i in range(len(word)): 
    messages_preset.append({
        "role": "assistant", 
        "content": response.choices[0].message.content
    })
    messages_preset.append({
        "role": "user", 
        "content": f"""
        生成规则：
        1. 任务：根据给定的背景，续写一段约100字的中文对话。
        2. 要求：
        a. 必须在提供的json对象的对话(dialog)字段中自然嵌入{word[i]}单词。
        b. 在生成对话前，检查json对象的dialog字段是否已包含{word[i]}单词。如果没有，则重新生成整个json对象。
        c. {word[i]}单词应与背景故事紧密结合。
        d. 对话中不得提示{word[i]}的中文意思。
        3. json对象结构：
        a. scene：描述对话发生的场景，应与对话内容中的场景一致。
        b. dialog：包含对话内容，需嵌入{word[i]}单词。
        c. npcName：对话中的非玩家角色名称，每三个场景更换一次。
        d. role_messages：提供可选的角色台词，可根据角色设定和背景氛围选择使用。
        e. field_messages：提供可选的场景描述，可根据对话场景选择使用。
        f. nextScene：指示下一个场景的编号，每次续写对话时需在上一个nextScene的基础上加1。
        g. options：包含三个选项，其中一个是正确翻译{word[i]}的选项，其余两个为错误翻译。正确选项位置随机。
        4. narrator：旁白，用于描述当前环境或提示操作。
        5. 主角：始终为“我”，不更换。
        6. 操作：选择正确选项后，对话将跳转至下一个场景。
        示例如下：
        {{
        "narrator":"你叫艾尔达，为了探索森林的秘密，你来到了一个古老且神秘的森林",
        "npcName":"森林守护者",
        "dialog":"你看起来真是个ambitious的冒险者啊，总想着探索这片古老的森林。但要小心，贪欲可是很危险的。",
        "scene":"峡谷",
        "options":[
                {{"text":"有野心的","isCorrect":true,"nextScene":1}},
                {{"text":"复杂的","isCorrect":false}},
                {{"text":"有充分的","isCorrect":false}}
        ]
        }}
        注意：只返回json
        """
    })
    response = client.chat.completions.create(
        model="glm-4-flash",  
        messages=messages_preset
    )
    print('休眠4秒钟')
    time.sleep(4)
    ## 检查
    messages_preset.append({
        "role": "assistant", 
        "content": response.choices[0].message.content
    })
    messages_preset.append({
        "role": "user", 
        "content": f"""
        你检查下符合我的规则吗?
        在对话(dialog)字段中自然嵌入{word[i]}单词。
        正确选项位置随机。
        如果符合，只返回json，如果不符合，返回修改后的json
        注意：无论是否符合，都只返回json，不要回复其他内容。
        """
    })
    response = client.chat.completions.create(
        model="glm-4-flash",  
        messages=messages_preset
    )
    print('单词' + word[i] + '修改后的json:' + response.choices[0].message.content)
    print('休眠4秒钟')
    time.sleep(4)
    print('现在进行到单词' + word[i] + '的生成')
    current_data = json.loads(response.choices[0].message.content.replace("```json", "").replace("```", "").strip())
    print(current_data)
    scenes_all.append({
    "narrator":str(current_data['narrator']),
    "npcName": str(current_data['npcName']),
    "npcDialog": str(current_data['dialog']),
    "npcAvatar": str(get_avatar_by_name(current_data['npcName'])),
    "backgroundVideo": "/videos/2077.mp4",
    "vocabulary": {
        "word": word[i],
        "options": current_data['options']
    }
    })
````
